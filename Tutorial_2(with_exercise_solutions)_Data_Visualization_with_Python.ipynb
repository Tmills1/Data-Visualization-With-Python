# Data Visualization
Data visualization is the graphical representation of data and information using visual elements such as charts, graphs, maps, and other visual formats. It aims to present complex datasets in a visually intuitive and accessible manner, making it easier to understand patterns, trends, relationships, and insights hidden within the data.

The primary goal of data visualization is to effectively communicate information and facilitate data-driven decision-making. By transforming raw data into visual representations, data visualization allows users to grasp complex concepts, identify patterns, and extract meaningful insights quickly. It can reveal correlations, outliers, distributions, and other patterns that may not be apparent in raw data alone.

# Anatomy of a Chart

Matplotlib enables us to create simple 2D plots with just a few commands. However, the true power of Matplotlib is in the underlying object structure.

### Understanding The Matplotlib Object Hierarchy

A Figure in matplotlib is divided into two different objects.

- `Figure object`
- `Axes object`

A Figure object is an instance of `matplotlib.figure.Figure`. It can contain one or more axes objects. `Figure is a container and we cannot plot data on it`.

Axes is the instance of `matplotlib.axes.Axes`. It is the `actual area on which our data is plotted`. One axes represents one plot inside a figure. Axes can contain other figure elements. All the other elements in our plot are also Python objects. The plot that we see as the output is the `Figure object` which is the top-level container for all the plotting elements. The Axes object holds the actual plot.

The image below represents the anatomy of a matplotlib figure:

  image.png



Key components of a matplotlib graph includes:

- **Figure**: The base of the graph, where all the other components are drawn.
- **Axis**: Contains the figure elements and sets the coordinate system.
- **Title**: The title gives the graph its name.
- **X-axis label**: The name of the x-axis, usually named with the units.
- **Y-axis label**: The name of the y-axis, usually named with the units.
- **Legend**: A description of the data plotted in the graph, allowing you to identify the curves and points in the graph.
- **Ticks and tick labels**: They indicate the points of reference on a scale for the graph, where the values of the data are. The labels indicate the values themselves.
- **Line plots**: These are the lines that are plotted with the data.
- **Markers**: Markers are the pictograms that mark the point data.
- **Spines**: The lines that delimit the border area of the chart.

## Line Graphs

A line graph, also known as a line chart, is a type of graph that displays data points connected by straight line segments. It is particularly useful for showing the trend or pattern of a variable over time or across different categories. Line graphs are commonly used in various fields, including business, economics, science, and social sciences.

Line graphs are especially effective for showing trends and patterns because the lines provide a continuous connection between data points. This allows you to see the direction and magnitude of change more easily. For example, if the line on the graph slopes upward, it indicates an increasing trend, while a downward slope indicates a decreasing trend. A horizontal line suggests no change over time or across categories.

To create a line graph, we need to:

- Pass the data into the `plt.plot()` [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).
- Display the plot using the `plt.show()` [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html).

The `plt.plot()` function generates a line graph by default. It takes in **two arrays of data of equal length**  these can be Python lists, pandas Series, NumPy arrays, etc. We will make a line plot of the number of COVID infection cases versus the month number. For this specific example, we will use two Python lists as inputs.

````python
month_number = [1, 2, 3, 4, 5, 6, 7]
new_cases = [9926, 76246, 681488, 2336640,
             2835147, 4226655, 6942042]

plt.plot(month_number, new_cases)
plt.show()
````

Notice the order of arguments in `plt.plot(month_number, new_cases)`: `month_number` comes first, followed by `new_cases`. **The array that comes first gives the x-coordinates, and the second array gives the y-coordinates.**

*Note: On the top left side of the graph, you can see a "1e6" sign  this is scientific notation. Matplotlib changes to scientific notation if one value on the axis needs to be one million or greater.*

### Customizing a Graph

Remember our line graph showing the evolution of new infection cases by month? In this section, you will learn how to customize a graph with matplotlib.

First, we will express the values of the y-axis in their raw numbers by removing the scientific notation. To achieve this we will make use of the `plt.ticklabel_format(axis, style)` [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ticklabel_format.html).

- The `axis` parameter defines which axis to configure  its arguments are the strings `x`, `y`, and `both`.

- The `style` parameter controls the notation style (plain or scientific). Its arguments are `sci`, `scientific`, and `plain`.

````python
plt.plot(month_number, new_cases)
plt.ticklabel_format(axis='y', style='plain')
plt.show()
````

The next thing we are going to do is to add a title to our graph. To achieve this we will use the `plt.title()` [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html).

````python
plt.plot(month_number, new_cases)
plt.ticklabel_format(axis='y', style='plain')
plt.title('New Reported Cases by Month  Globally')
plt.show()
````

Remember that the x-axis shows the month number, and the y-axis shows the number of infection cases. We can show this on our graph by adding a label to each axis  `a y-label and an x-label`. To add axis labels, we use `plt.xlabel()` and `plt.ylabel()`.

Adding a title and axis labels is always a good visualization practice.

````python
plt.plot(month_number, new_cases)
plt.ticklabel_format(axis='y', style='plain')
plt.title('New Reported Cases By Month (Globally)')
plt.xlabel('Month Number')
plt.ylabel('Number Of Cases')
plt.show()
````

## Subplots

When analyzing data you might want to compare multiple plots placed side-by-side. Matplotlib provides a convenient method called subplots to do this. In Matplotlib, a subplot refers to a grid of plots or axes within a single figure. It allows you to create multiple plots in a single figure, arranged in a grid-like structure, which is useful for comparing and visualizing different aspects of your data. Think of a figure as a canvas that holds multiple plots.

This is an example grid layout:

image.png

When using `plt.subplots()` to create a new figure, it is best practice to specify the figure size using the `figsize` parameter. You are expected to pass in a tuple corresponding to `(width, height)`.

Let's attempt to visualize a 1 by 2 grid plot (i.e two plots side by side). We will make empty plots using this template:

````python
# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,7)) # sharex='col', sharey='row'
plt.show()
````

Now that you are familiar with subplotting principles. let's plot the cummulative reported cases forr france besides that of the UK. To do this, we will need to specify the parameters `nrows` and `ncols`, in our case 1 and 2 respectively. This will create an empty plot with the desired template.

Note that plt.subplot will return  the figure `fig` and the empty axes template `ax`. However, in order to fill these empty axes with our desired line plots, we will index the axes as though it was a 1 by 2 matrix.

````python
# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))
ax[0].plot(UK['Date_reported'], UK['Cumulative_cases'])
ax[1].plot(france['Date_reported'], france['Cumulative_cases'])

#Let's add their respective titles.
ax[0].set_title("UK cummulative reported COVID-19 cases")
ax[1].set_title("France cummulative reported COVID-19 cases")

plt.show()
````

## Correlation Heatmap

A correlation heatmap is a graphical representation of the correlation matrix, which displays the correlation coefficients between multiple variables as a color-coded matrix. It provides a visual way to understand the relationships and dependencies among variables in a dataset.


A correlation heatmap displays the correlation coefficients between variables as a color-coded matrix. Each cell represents the correlation coefficient, which measures the strength and direction of the linear relationship between two variables. The coefficient ranges from -1 to +1: -1 indicates a perfect negative correlation, +1 indicates a perfect positive correlation, and 0 indicates no correlation. The color of each cell reflects the coefficient, typically using a color gradient. The colors can range from cool colors like blue for negative correlations, through neutral colors like white for no correlation, to warm colors like red for positive correlations.

To visualize a correlation heatmap, we will need the correlation matrix. A correlation matrix. A correlation matrix is a square matrix that displays the correlation coefficients between pairs of variables in a dataset. It provides a comprehensive view of the relationships and dependencies among multiple variables. In a correlation matrix, each cell represents the correlation coefficient between two variables

We can obtain this using the `.corr()` function in pandas.

````python
# Create the correlation matrix
correlation_matrix = bike_sharing.corr()
````

Next, we will generate a correlation heatmap using Matplotlib. First, we will creates a figure and axes for plotting using `plt.subplots()`, then we will proceed to use the `imshow` [function](https://www.geeksforgeeks.org/matplotlib-pyplot-imshow-in-python/) to display the correlation matrix as a heatmap with a `cool-to-warm` color gradient.

The next step will then be to add a colorbar which indicates the correlation values. The x and y axis labels are set using the column names of the **correlation matrix** in the dataset. We will also x-axis tick labels are rotated for better readability.

````python
import numpy as np

# Create a figure and axes
fig, ax = plt.subplots()

# Create the heatmap
heatmap = ax.imshow(correlation_matrix, cmap='coolwarm')

# Add colorbar
cbar = fig.colorbar(heatmap)

# Set x and y axis labels
ax.set_xticks(np.arange(correlation_matrix.shape[0]))
ax.set_yticks(np.arange(correlation_matrix.shape[1]))

#Pass in the column names of the correlation matrix as the ticklabels
ax.set_xticklabels(correlation_matrix.columns)
ax.set_yticklabels(correlation_matrix.columns)

# Rotate x-axis tick labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Set the title
ax.set_title("Correlation Heatmap")

# Display the heatmap
plt.show()
````

Overall, a correlation heatmap provides a concise summary of the pairwise relationships between variables, allowing you to gain insights into the data and make informed decisions in data analysis and modeling tasks.

Now, let's create a bar plot representing the average number of casual bikers who rented bikes on each weekday from our dataset. To achieve this we will group the dataset by `weekday` and then compute the mean of each group. The resulting dataframe will correspond to the average number of causal bikers per weekday

````python
# Group by 'weekday' and calculate the average of 'casual' and 'registered'
weekday_averages = bike_sharing.groupby("weekday")[['casual','registered']].mean()
# Convert the grouped DataFrame to a normal DataFrame
weekday_averages = weekday_averages.reset_index()
weekday_averages
````

### Customizing Bar Plots

Notice that we now have seven rows in the `weekday_averages` dataframe, and it's a bit harder to compare all the values. We can use a bar plot to help us get a visual estimate of the amount of bikes rented by casual bikers on each weekday.

````python
plt.bar(weekday_averages['weekday'], weekday_averages['casual'])
````

The first bar (Sunday) and the last one (Saturday) are more than twice as tall as all the other bars. This shows that casual users rent more bikes on the weekends.

If we look at the x-tick labels, we only see numbers (from 0 to 6). If we wanted to add day names instead, we can use the `plt.xticks()` function. The ticks parameter takes in the x-coordinates, and the labels parameter takes in the corresponding new labels. Additionally, to prevent our tick lables from overlapping, one thing we can do is leverage the `rotation` parameter.

````python
plt.bar(weekday_averages['weekday'], weekday_averages['casual'])
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6],
          labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday',
                 'Thursday', 'Friday', 'Saturday'],
          rotation=30)
plt.show()
````

To prevent overlapping xtick labels, we can also flip all yhe bars horizontally. This is called a **horizontal bar plot**.  The average numbers of bikes rented are now on the x-axis, and the day names are on the y-axis. And we're now talking about bar widths instead of bar heights.

To generate this type of barplot, we use the `plt.barh(y, width)` [function](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.barh.html) along with the `plt.yticks()` function to change the y-tick labels:

````python
plt.barh(weekday_averages['weekday'], weekday_averages['casual'])
plt.yticks(ticks=[0, 1, 2, 3, 4, 5, 6],
          labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday',
                 'Thursday', 'Friday', 'Saturday'])
plt.show()
````

## Histogram

A histogram is a graphical representation of the distribution of a dataset. It is a way to visualize the frequency or count of observations falling into different intervals or bins. In the context of histograms, a bin refers to a specific interval or range into which the data is divided for analysis and visualization. Each bin represents a subset of values from the dataset that fall within a particular range.

The x-axis of a histogram represents the range of values in the dataset, divided into several equal-width bins. The y-axis represents the frequency or count of observations falling into each bin.

Significance of Histograms: Histograms are commonly used to understand the shape, central tendency, and spread of a dataset. They provide insights into the distribution of data, such as whether it is symmetric, skewed, or bimodal. Histograms can also reveal the presence of outliers or gaps in the data.

Let's attempt to generate a frequency table for the `cnt` variable.

````python
bike_sharing['cnt'].value_counts()
````

 The `cnt` column (which describes the total number of bike rentals)  has 696 unique values (notice the value at the end of the truncated table). Thus, it's almost impossible to distinguish any pattern.

 To solve this problem, we can group the unique values into equal intervals. Below, we group the table into ten equal intervals by using the `bins=10` argument inside the `.value_counts()` method

````python
bike_sharing['cnt'].value_counts(bins=10).sort_index()
````

The unique values are now number intervals. (4368.0, 5237.2] is a number interval. The ( character indicates that the starting number is not included, while the ] indicates that the ending number is included. The interval (4368.0, 5237.2] contains all numbers greater than 4368.0 and less than or equal to 5237.2.

This is called a `grouped freqeuncy table`. The best way to visualize a grouped frequency table is by using a histogram. To generate a histogram for our table above, we can use the `plt.hist()` [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

````python
plt.hist(bike_sharing['cnt'])
plt.show()
````

When using the `plt.hist()` function in matplotlib, only the column containing the data (in this case, the `cnt` column) needs to be provided, not the grouped frequency table. The function automatically generates a grouped frequency table with equal intervals based on the data. It then plots a bar for each interval, where the height of each bar represents the frequency of that interval. **Histograms are a variation of bar plots, where the bars are contiguous without any gaps between them**.

## Visualizing Distributions with Histograms


Different types of distributions that can be observed from a histogram include:

- **Uniform Distributio**n: In a uniform distribution, the data is evenly distributed across the range of values, resulting in a flat and constant height of bars in the histogram.

- **Normal Distribution**: A normal distribution exhibits a symmetric bell-shaped curve in the histogram. The majority of the data points are clustered around the mean, with fewer points towards the tails.

- **Skewed Distribution**: Skewed distributions are asymmetric, with a longer tail on one side. If the tail extends to the `right` (positive values), it is called a `right-skewed` or `positively skewed` distribution. Conversely, if the tail extends to the `left` (negative values), it is a `left-skewed` or `negatively skewed` distribution.

- **Bimodal Distribution**: A bimodal distribution displays two distinct peaks or modes in the histogram. It suggests the presence of two separate groups or populations within the data.

- **Multimodal Distribution**: A multimodal distribution has more than two peaks or modes, indicating the presence of multiple groups or populations with different characteristics.

# Boxplot (Whisker plot)

A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, including the minimum, first quartile (25th percentile), median (second quartile or 50th percentile), third quartile (75th percentile), and maximum. Additionally, it provides information about potential outliers in the data.

A boxplot consists of a rectangle (the box) and two lines (the whiskers) that extend from the box. The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The median is typically marked within the box. The whiskers extend from the box to the minimum and maximum values within a specified range. They can also be drawn to the nearest data point within a certain distance, depending on the specific conventions or definitions used.

Potential outliers, which are values that lie significantly outside the range of the rest of the data, can be represented as individual points or dots on the plot. Outliers are often defined as values that fall below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR`. However, the specific criteria for identifying outliers can vary.

To make a boxplot, we first define a figure using the plt.figure() function. Additionally, we will pass in the figure size. Then `plt.boxplot()` function is used to visualize the bocplot.

````python
fig = plt.figure(figsize =(10, 7))

# Creating plot
plt.boxplot(bike_sharing['hum'])

# show plot
plt.show()
````

# Pie Chart

A pie chart is a circular graphical representation that displays data as a divided pie-like structure. The chart is divided into sectors, and each sector represents a proportion or percentage of the whole dataset. The size of each sector corresponds to the magnitude of the data it represents.

Pie charts are particularly useful for illustrating parts-to-whole relationships and displaying categorical data. The entire pie represents the whole dataset, and each sector represents a specific category or subgroup within that dataset. The angle of each sector is proportional to the relative size or proportion of the data it represents.

To create a pie chart in matplotlib, we make use of the `plt.pie()` [function](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html). The piec chart can only be created for a categorical variable. To create a pie chart plot, you need two things:
- data: A list of frequencies of the categories in the categorical column.
- label: A list of categories in the categorical column.

However, this doesn't means that the only a frequency table is required to plot a pie chart. You can plot a piechart of any quantity that is related to the groups within a categorical variable. For instance, the total number of bikes rented by casual users for each weekday in June.


````python
# Create a random dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)

# show plot
plt.show()
````

### Customizing a Pie Chart

A pie chart can be customized on the basis several aspects. Let' try to modify the previous plot.

Explode a Piechart: To "explode" a pie chart means to visually separate or highlight one or more slices of the pie chart by pulling them out from the rest of the chart. This creates a visual effect where the selected slices are visually distinct or appear to be "exploded" from the center of the pie. presented.

To achieve this, we will define a list of decimal (between 0 and 1) whose length is equal to the length of `weekday` categories.
If you intend that a category be exploded you will specify a decimal, otherwise 0. The closer the value is to 1, the larger the explosion.

````python
explode = [0.1, 0, 0, 0, 0,0,0]
plt.pie(x=data, labels= weekday_list, explode=explode, radius=1.5)
plt.show()
````

Customize a Piechart with shadow: specify the shadow parameter as True

````python
plt.pie(x=data, labels= weekday_list, explode=explode, radius=1.5, shadow=True)
plt.show()
````

Customize colors in pie chart: we will pass in a list of colors into the `plt.pie()` function. It is important to note that the list of colors must be equal to the number of categories in `weekday` column.

````python
colors = ['indianred', 'deeppink', 'gold', 'mediumorchid', 'palegreen','blue','beige']
plt.pie(x=data, labels= weekday_list, explode=explode, colors=colors, radius=1.5, shadow=True)
plt.show()
````

Label the Pie with percentages: To achieve this, we will pass in another parameter called `autopct`. The value for this parameter will be the string formatting rule for the percentages. In our example, we will express the percentages in 1 decimal place, for this reason, we will use the `%.1f%%` rule.

````python
plt.pie(x=data, labels= weekday_list, explode=explode, colors=colors, radius=1.5, shadow=True, autopct='%.1f%%')
plt.show()
````

Pie chart with legend: We will utilize the `plt.legend()` function for this. We will pass the legend title and the intenddistanceed position for the legend as `loc`. Additionally, we will define another parameter `label` with a value of `None`. This will remove the labels surrounding the pie chart.

````python
plt.figure(figsize=(10,8))
plt.pie(x=data, labels= weekday_list, explode=explode, colors=colors,
        radius=1.2, shadow=True, autopct='%.1f%%',labeldistance=None)

plt.legend(title="Weekday", loc="upper right")
# other values of `loc`: upper left, upper right, lower left, lower right, upper center, lower center, center left, center right, center, best
plt.show()
````
